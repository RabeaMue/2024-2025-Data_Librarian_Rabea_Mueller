## OpenAPC data analysis using pandas

```python
import urllib.request
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/refs/heads/master/data/fuberlin/APC_FU_Berlin_2015.csv"
apc_file = "openapc.csv"
```

```python
# Retrieving data
urllib.request.urlretrieve(url, apc_file)
```

```python
# Load data into DataFrame
apcs = pd.read_csv(apc_file)
```

```python
# Create new DataFrame that contains only the two selected columns
apcs_publisher_and_spending = apcs[ ["publisher", "euro"]  ]
```

```python
apcs_publisher_and_spending
```

```python
# Group data according to one column. This per se is not so useful ...
apcs_publisher_and_spending.groupby("publisher")
```

```python
# but the DataFrameGroupBy object has several useful method e.g. sum
# to sum up the values of all euro value of all entries that belong to
# one publisher
apcs_publisher_and_spending.groupby("publisher").sum()
```

```python
apcs_publisher_and_spending.groupby("publisher").mean()
```

```python
apcs_publisher_and_spending.groupby("publisher").mean().columns
```

```python
apcs_publisher_and_spending.groupby("publisher").mean().index
```
